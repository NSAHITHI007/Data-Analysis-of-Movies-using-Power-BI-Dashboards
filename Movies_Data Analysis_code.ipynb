# Missing Values in the dataset

```python
missing_values = tmdb_movies_data.isnull().sum()
print("The Missing Values in Each Column are:\n", missing_values)
```

# Dealing with the missing values in the dataset

```python
columns_to_drop = ['backdrop_path', 'homepage', 'imdb_id', 'overview', 
                   'poster_path', 'tagline', 'production_countries', 
                   'spoken_languages']
tmdb_movies_data.drop(columns=columns_to_drop, inplace=True)
tmdb_movies_data.dropna(subset=['release_date'], inplace=True)
tmdb_movies_data['genres'].fillna('Unknown', inplace=True)
tmdb_movies_data['production_companies'].fillna('Unknown', inplace=True)
tmdb_movies_data['release_date'] = pd.to_datetime(tmdb_movies_data['release_date'])
tmdb_movies_data['release_year'] = tmdb_movies_data['release_date'].dt.year
current_year = pd.to_datetime('today').year
tmdb_movies_data = tmdb_movies_data[tmdb_movies_data['release_year'] <= current_year]
tmdb_movies_data['title'] = tmdb_movies_data['title'].str.strip()
tmdb_movies_data['title'] = tmdb_movies_data['title'].str.title()
tmdb_movies_data['title'] = tmdb_movies_data['title'].str.replace(r'[^A-Za-z0-9 ]+', '', regex=True)
tmdb_movies_data = tmdb_movies_data.sort_values('title', ascending=True)
print(tmdb_movies_data.head())
print("\nThe Remaining Missing Values in the dataset are:\n", tmdb_movies_data.isnull().sum())
tmdb_movies_data.to_csv('modified_tmdb_movies_data.csv', index=False)
```

# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
tmdb_movies_data['release_month'] = tmdb_movies_data['release_date'].dt.month
tmdb_movies_data['release_year'] = tmdb_movies_data['release_date'].dt.year
monthly_revenue = tmdb_movies_data.groupby('release_month')['revenue'].mean()
plt.figure(figsize=(8, 6))
sns.barplot(x=monthly_revenue.index, y=monthly_revenue.values)
plt.title('Revenue of Movies by their Releasing Month')
plt.xlabel('Month')
plt.ylabel('Average Revenue')
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
from itertools import chain
import matplotlib.pyplot as plt
import seaborn as sns
tmdb_movies_data['genres'] = tmdb_movies_data['genres'].str.split(', ')
exploded_genres = tmdb_movies_data.explode('genres')
genre_metrics = exploded_genres.groupby('genres')['revenue', 'popularity', 'vote_average'].mean().reset_index()
top_genres_by_revenue = genre_metrics.sort_values('revenue', ascending=False).head(10)
top_genres_by_popularity = genre_metrics.sort_values('popularity', ascending=False).head(10)
top_genres_by_vote = genre_metrics.sort_values('vote_average', ascending=False).head(10)
def create_bar_plot(df, y_col, x_col, title, x_label, ax):
    sns.barplot(x=x_col, y=y_col, data=df, ax=ax)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel('Genre')
fig, axes = plt.subplots(3, 1, figsize=(10, 15))
create_bar_plot(top_genres_by_revenue, 'genres', 'revenue', 'The Topmost 10 Genres by Average Revenue', 'Average Revenue (USD)', axes[0])
create_bar_plot(top_genres_by_popularity, 'genres', 'popularity', 'The Topmost 10 Genres by Average Popularity', 'Average Popularity', axes[1])
create_bar_plot(top_genres_by_vote, 'genres', 'vote_average', 'The Topmost 10 Genres by Average Vote Average', 'Average Vote Average', axes[2])
plt.subplots_adjust(hspace=0.4)  
plt.tight_layout()
plt.show()
```

# Creating Models

# Building Linear Regression Model

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
numeric_data = tmdb_movies_data.select_dtypes(include=[np.number]).drop(['id'], axis=1)
X = numeric_data.drop('vote_average', axis=1)
y = numeric_data['vote_average']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)
mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = mse_lr ** 0.5
r2_lr = r2_score(y_test, y_pred_lr)
print(f'The Linear Regression Model has got  RMSE: {rmse_lr}')
print(f'The Linear Regression Model has got R^2 Score: {r2_lr}')
```

# Building Random Forest Regressor Model

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
numeric_columns = tmdb_movies_data.select_dtypes(include=['int64', 'float64']).columns
numeric_data = tmdb_movies_data[numeric_columns]
numeric_data = numeric_data.drop(['id'], axis=1)
X = numeric_data.drop('vote_average', axis=1)
y = numeric_data['vote_average']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
rf_model = RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)
print(f'The Random Forest Model got RMSE: {rmse}')
print(f'The Random Forest Model got R^2 Score: {r2}')
```
